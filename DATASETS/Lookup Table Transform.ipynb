```python
import csv
import pandas as pd
import numpy as np

source = 'Education Counts Summary Tables/' 
fname = source+'Provider Summary Tables 2016 LOOKUP INTERPOLATED'
col_name = 'Student Enrolments'
transform_fname = source+'Provider Summary Tables 2016 LOOKUP INTERPOLATED REFORMATTED'
```

```python
file = pd.read_csv(fname+'.csv', parse_dates=['Month'], dayfirst=True)
file
```

```python
regions = list(file)
del regions[0]

transformed = pd.DataFrame()

for region in regions:
    region_dataframe = file['Month'].to_frame()
    region_dataframe['Region']=region
    region_dataframe[col_name]=file[region]
    transformed = pd.concat([transformed, region_dataframe])
    
transformed.to_csv(transform_fname+'.csv')
transformed
```
